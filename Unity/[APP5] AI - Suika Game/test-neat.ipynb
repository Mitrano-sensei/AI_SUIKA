```python
import neat
import numpy as np
import random
from mlagents_envs.environment import UnityEnvironment




# Définir une fonction d'évaluation (fitness function) pour évaluer les performances des individus
def eval_genomes(genomes, config):
    global ge, nets;
    env.reset()
    env.step()
    done = False
    tracked_agent = -1
    ge=[]
    nets=[]
    for genome_id, genome in genomes:
        genome.fitness = 0
        ge.append(genome)
        net = neat.nn.FeedForwardNetwork.create(genome, config)
        nets.append(net)
        
        # Votre logique d'évaluation ici

    while not done:
       
        for i in range(0,len(list(env.behavior_specs.keys()))):
            decision_steps, terminal_steps = env.get_steps(list(env.behavior_specs.keys())[i])
            

            if (i == 0): # Tracking steps for first agent only
                if tracked_agent == -1 and len(decision_steps) >= 1:
                    tracked_agent = decision_steps.agent_id[0]
                if tracked_agent in terminal_steps:
                   done = True
                   break

            ge[i].fitness=decision_steps.reward[0]+ge[i].fitness
            # print(list(env.behavior_specs.keys())[i])
            spec = env.behavior_specs[list(env.behavior_specs.keys())[i]]
            

            output = nets[i].activate(decision_steps.obs[0][0])
            # print(output)

            # Récupérer les observations des agents pour un comportement spécifique
            pos=np.array([[(output[0])]]) 
            # onJoue=np.array([[random.random() > .5]])
            onJoue=np.array([[output[1]>0]])
            
            

            creationAction(pos,onJoue,spec, i)
            # print(decision_steps.reward[0])
            
        env.step()
    # for i in range (0,len(ge)):
    #     for j in range (0,10+1):
    #         #Il faut récupérer les inputs
    #         #Faut réussir à appeler la fonction CollectObservations du GameManager

    #         #Il mettre les input dans le réseau de neuronnes
    #         output = nets[i].activate() #Prend en param les inputs

    #         #Lance l'action voulu par l'output 
    #         #On doit réussir à lancer la fonction OnActionReceived du gameManager


    #         #On récupère Reward et on met à jour fitness


        

def run(path="config.txt"):
    # Charger la configuration par défaut de NEAT
    config = neat.Config(neat.DefaultGenome, neat.DefaultReproduction,
                        neat.DefaultSpeciesSet, neat.DefaultStagnation,
                        path)  # Vous pouvez spécifier un fichier de configuration personnalisé

    # Créer une population initiale
    p = neat.Population(config)

    # Ajouter des rapports pour afficher des statistiques pendant l'évolution
    p.add_reporter(neat.StdOutReporter(True))  # Affiche des statistiques dans la console
    p.add_reporter(neat.StatisticsReporter())   # Collecte des statistiques sur la population

    # Évaluer la population sur 10 générations
    winner = p.run(eval_genomes, 100)

    # Utiliser le gagnant (meilleur individu) pour votre tâche spécifique
    return winner;


def creationAction(pos,onJoue,spec, id):
    action=spec.action_spec.empty_action(1)
    action.add_continuous(pos)

    action.add_discrete(onJoue)
    env.set_actions(list(env.behavior_specs.keys())[id], action)


try:
    env.close()
except:
    pass

env = UnityEnvironment(file_name="./build/SuikaAI",worker_id=1)
res=run(path="config.txt")
env.close()
```
